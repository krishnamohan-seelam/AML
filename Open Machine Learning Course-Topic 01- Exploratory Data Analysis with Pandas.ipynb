#### Open Machine Learning Course-Topic 01- Exploratory Data Analysis with Pandas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
sns.set_palette(sns.color_palette("muted", 15))
```

```python
filename = r'.\data\telecom_churn.csv'
telecom_df = pd.read_csv(filename)
telecom_df.head()
```

```python
print(telecom_df.info())
```

```python
print(telecom_df.columns)
```

```python
print(telecom_df.describe())
```

#### Change data type of attribute

```python
telecom_df['Churn']=telecom_df['Churn'].astype('int64')
print("-"*40)
print("value count")
print(telecom_df['Churn'].value_counts())
print("-"*40)
print("normalize value count")
print(telecom_df['Churn'].value_counts(normalize=True))
```

####  To see statistics on non-numerical features, one has to explicitly indicate data types of interest in the include parameter

```python
telecom_df.describe(include=['object','bool'])
```

#### Sorting 

```python
telecom_df.sort_values(by=['Churn','Total day charge'] ,ascending =[True,False]).head()
```

####  What are average values of numerical variables for churned users?

```python
telecom_df[telecom_df['Churn']==1].mean()
```

####  How much time (on average) do churned users spend on phone during daytime?

```python
telecom_df[telecom_df['Churn']==1]['Total day minutes'].mean()
```

#### What is the maximum length of international calls among loyal users (Churn == 0) who do not have an international plan?

```python
telecom_df[ (telecom_df['Churn']==1)&(telecom_df['International plan']=='No')]['Total intl minutes'].mean()
```

#### DataFrames can be indexed by column name (label) or row name (index) or by the serial number of a row. The loc method is used for indexing by name, while iloc() is used for indexing by number.

```python
telecom_df.loc[0:5,'State':'Area code']
```

```python
telecom_df.iloc[0:5,0:3]
```

####  Applying Functions to Cells, Columns and Rows
#### The apply method can also be used to apply a function to each line. To do this, specify axis=1. Lambda functions are very convenient in such scenarios. 

```python
telecom_df.apply(np.max)
```

####  select all states starting with W

```python
telecom_df[telecom_df['State'].apply(lambda x :x[0]=="W")].head()
```

#### The map method can be used to replace values in a column by passing a dictionary of the form {old_value: new_value} as its argument:

```python
international_plan = {'No' : False, 'Yes' : True} 
telecom_df['International plan'] =telecom_df['International plan'].map(international_plan)
telecom_df.head()
```

#### Grouping

```python
columns_to_show = ['Total day minutes', 'Total eve minutes', 'Total night minutes']
telecom_df.groupby(['Churn'])[columns_to_show].describe(percentiles=[])
```

```python
telecom_df.groupby(['Churn'])[columns_to_show].agg([np.mean,np.std,np.max,np.min,len])
```

#### Summary 

```python
pd.crosstab(telecom_df['Churn'],telecom_df['International plan'])
```

```python
pd.crosstab(telecom_df['Churn'],telecom_df['Voice mail plan'],normalize =True)
```

#### Pivot tables

```python
telecom_df.pivot_table(['Total day calls', 'Total eve calls', 'Total night calls'], ['Area code'], aggfunc='mean')
```

#### DataFrame transformations

```python
telecom_df['Total charge'] = (telecom_df['Total day charge'] + telecom_df['Total eve charge'] +  
                              telecom_df['Total night charge'] + telecom_df['Total intl charge'])
```

```python
telecom_df['Total charge'].head()
```

```python
telecom_df['Total_charge'] =  (telecom_df [['Total day charge','Total eve charge','Total night charge',
                                            'Total intl charge']].sum(axis=1))
```

```python
telecom_df['Total_charge'].head()
```

#### Deleting columns from dataframe

```python
telecom_df.drop(['Total charge'], axis=1, inplace=True) 
```

```python
# some imports and "magic" commands to set up plotting 
%matplotlib inline
plt.rcParams['figure.figsize'] = (8, 6)
sns.countplot(x='International plan', hue='Churn', data=telecom_df)
```

```python
sns.countplot(x='Customer service calls', hue='Churn', data=telecom_df)
```

```python
telecom_df['Many_service_calls'] = (telecom_df['Customer service calls'] > 3).astype('int')
pd.crosstab(telecom_df['Many_service_calls'], telecom_df['Churn'], margins=True)
sns.countplot(x='Many_service_calls', hue='Churn', data=telecom_df)
```

```python
pd.crosstab(telecom_df['Many_service_calls'] & telecom_df['International plan'] , telecom_df['Churn'])
```
